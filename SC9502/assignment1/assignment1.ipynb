## Question 1

Computing can generally be measured using the number of *floating point operations* performed in a *second*, known as "FLOPS". This is distinct from the number typically associated with CPUs, which is the clock frequency. The difference is that the clock frequency measures the number of cycles per second (i.e. times the CPU can execute an instruction), whereas the FLOPS measures the number of floating point operations that can be performed in one second, where multiple floating point operations can per performed per cycle.

We can compare the various computing platforms from the past and present using this generic measure. The comparison will ignore the computing power of graphics processors, and only compare the central processing units (CPUs).

Supercomputers in the 1990s took a massive leap by incorporating thousands of processors compared to the mere 4 CPUs of the previous decade's flagship, the Cray-2 (Anderson, 1988). Two notable examples of supercomputers from the 1990s are: the Intel Paragon XP/S (Esser, 1993), where the XP/S 140 achieved a peak performance of 143 GFLOPS and claimed the number one position in 1994; and the *Numerical Wind Tunnel* built in Japan with a peak performance of 170 GFLOPS ($10^9$ FLOPS) (Hirose, 1989), snatching number one position from the XP/S until it was dethroned in 1995 (TOP500.org).

The FLOPS measurements of supercomputers of this era was typically determined from the LINPACK benchmark test, which is based on evaluating the performance of solving a general dense matrix problem (Dongarra, 2003), and is still maintained today. The computer used to write this assignment achieved a maximum measurement of 239 GFLOPs on the benchmark (using an Intel(R) Core(TM) i7-6800K CPU @ 3.40GHz).

Another benchmark is called the Whetstone benchmark, developed in 1976 and the first one made specifically for benchmarking.

### Question 3-16

Choice of method is `LinearRegression` from the skit-learn library. This is a very straightforward way to get the linear regression parameters and plot them.

````python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

model = LinearRegression()

fig = plt.figure()
ax1 = fig.add_subplot(111)

x1c = sm.add_constant(x1)
model1 = sm.OLS(y1, x1c)
results1 = model1.fit()

x2c = sm.add_constant(x2)
model2 = sm.OLS(y2, x2c)
results2 = model2.fit()

b1, m1 = results1.params
b2, m2 = results2.params

ax1.scatter(x1, y1, s=1, label="GROtesk MACabre and Sinister")
model.fit(x1.reshape((-1, 1)), y1)
ax1.plot(
    x1,
    [m1 * x + b1 for x in x1],
    linestyle="solid",
    linewidth=1,
    label="Fit to GROtesk MACabre and Sinister",
    c="#000e7d",
)

ax1.scatter(x2, y2, s=1, marker="x", label="Grunge ROck MAChoS")
model.fit(x2.reshape((-1, 1)), y2)
ax1.plot(
    x2,
    [m2 * x + b2 for x in x2],
    linestyle="solid",
    linewidth=1,
    label="Fit to Grunge ROck MAChoS",
    c="#6b2d00",
)

# ax1.plot(x2,y2, label='Grunge ROck MAChoS')
ax1.legend()

ax1.set_title("GROMACS Energies")
ax1.set_xlabel("Time (ps)")
ax1.set_ylabel("Total Energy (kJ/mol)")
````

### Question 3-18

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from fitter import Fitter, get_common_distributions, get_distributions

data = np.abs(y2 - np.fromiter((m2 * x + b2 for x in x2), float))

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.hist(data, bins=20, rwidth=0.9, label="Grunge ROck MAChoS")
ax1.legend()
ax1.set_xlabel("Deviation in Energy from fit (kJ/mol)")
ax1.set_ylabel("Count")
ax1.set_title("Distribution of deviations from linear regression fit")

f = Fitter(
    data,
    distributions=[
        "gamma",
        "lognorm",
        "beta",
        "norm",
        "binomial",
        "expon",
        "rayleigh",
        "chi2",
    ],
)
fit = f.fit()
best = f.get_best(method="sumsquare_error")
````

````python
summary = f.summary()
best
````

````python
xmax = data.max()
xx = np.linspace(0, xmax, 50)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.hist(data, bins=40, rwidth=0.9, label="Grunge ROck MAChoS")
ax1.legend()
ax1.set_xlabel("Deviation in Energy from fit (kJ/mol)")
ax1.set_ylabel("Count")
ax1.set_title("Distribution of deviations from linear regression fit")
ax1.plot(xx, stats.beta.pdf(xx, **best['beta']))
````

While the beta distribution displays the best fit overall from the sum of squares error:
$$
\sum^n_{i=1} (y_i - \bar{y})^2,
$$
it still does not show the same qualitative behaviour. In particular, there are two distinct peaks, one very close to zero and onother near 0.0025. This is not captured by any distribution.

Therefore, the data cannot be fitted to any of the simple distributions.

### 2.1

The probability density is given by taking each value and dividing by the number of experiments conducted, $N$, where $N = 175$. The mean and standard deviation are: $\mu = \langle \Lambda \rangle = 0.449$, $\sigma = 0.204$. The mean of the square value is: $\langle \Lambda^2 \rangle = 0.243$.

The PDF table:
 
| $\Lambda$ | $P(\Lambda)$ |
|--------------|-------------|
| 0.0-0.2 | 0.08571429
| 0.2-0.4 | 0.37142857
| 0.4-0.6 | 0.31428571
| 0.6-0.8 | 0.17142857
| 0.8-1.0 | 0.05714286

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

fig = plt.figure()
ax1 = fig.add_subplot(111)
y = np.array([15, 65, 55, 30, 10])
x = np.arange(0.1, 1, .2)
N = np.sum(y)

mean = np.sum(x * y) / N
mean2 = np.sum(y * (x) ** 2) / N
std = np.sqrt(np.sum(y * np.power(x - mean, 2)) / N)

ax1.bar(x, y / N, width=0.19)

# x_norm = np.arange(0.0, 1, .05)
# ax1.plot(x_norm, stats.norm.pdf(x_norm, mean, std))
ax1.set_ylabel("Probability")
ax1.set_xlabel(r"$\Lambda$")

print(N)
print(mean, mean2, std)
```
